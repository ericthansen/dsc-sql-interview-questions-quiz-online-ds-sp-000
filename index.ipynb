## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

A primary key of a table is a unique identifier for a given row of that table.  Foreign keys are columns of other tables that may not be unique but establishes a relationship between a given table and another table.

## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Any variable that is being aggregrated should have an accompanying variable by which it is grouped.  I.e. the group by variable(s) determine the collections by which the aggregations are grouped.  
E.g., one could aggregate total $ spent on orders by customer, or by salesperson, or all $ spent for entire company; each would entail a different way of grouping $ spent.
